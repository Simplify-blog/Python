```python
#forループで複数グラフを描く際にはaxes.flattenが便利
#axes.flattenを使う場合
import numpy as np
import matplotlib.pyplot as plt
```

```python
#forループで複数グラフを描く際にはaxes.flattenが便利
#flattenあり
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
n = 100
np.random.seed(123)
ax = ax.flatten()
for i in range(len(ax)):
    #データの生成
    x = np.random.rand(n)
    y = 2 * x + np.random.rand(n)
    #散布図
    ax[i].scatter(x, y)
    ax[i].set_xlabel('x_feature', fontsize=14)
    ax[i].set_ylabel('y_feature', fontsize=14)
plt.tight_layout()
print('ax', ax)
print('ax.shape:', ax.shape)
print('type(ax)', type(ax))
fig.savefig('flatten.png', bbox_inches='tight')
```

```python
#flattenなし
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
n = 100
np.random.seed(123)
nrows, ncols = ax.shape
for i in range(nrows):
    for j in range(ncols):
        x = np.random.rand(n)
        y = 2 * x + np.random.rand(n)
        ax[i, j].scatter(x, y)
        ax[i, j].set_xlabel('x_feature', fontsize=14)
        ax[i, j].set_ylabel('y_feature', fontsize=14)
plt.tight_layout()
print('ax:', ax)
print('ax.shape:', ax.shape)
print('type(ax):', type(ax))
```
